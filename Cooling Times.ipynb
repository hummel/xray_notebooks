```python
%matplotlib inline
import os
import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt
mpl.rc('figure', figsize=(16,12))
mpl.rc('font', size=20.)
mpl.rc('font', family='serif')
mpl.rc('text', usetex=True)
```

```python
yr = 3600 * 24 * 365.25
gs = 500
s = 1
cool = plt.cm.Blues_r
warm = plt.cm.autumn
```

# At Sink Formation

```python
V = np.load(os.getenv('HOME')+'/sim/chemistry/stampede/vanilla/cooltime_0355.npz')
J0 = np.load(os.getenv('HOME')+'/sim/chemistry/stampede/XR_sfr_1e-3/cooltime_0269.npz')
J10 = np.load(os.getenv('HOME')+'/sim/chemistry/stampede/XR_sfr_1e-2/cooltime_0201.npz')
J100 = np.load(os.getenv('HOME')+'/sim/chemistry/stampede/XR_sfr_1e-1/cooltime_0200.npz')

hV = np.load(os.getenv('HOME')+'/sim/chemistry/stampede/vanilla/heattime_0355.npz')
hJ0 = np.load(os.getenv('HOME')+'/sim/chemistry/stampede/XR_sfr_1e-3/heattime_0269.npz')
hJ10 = np.load(os.getenv('HOME')+'/sim/chemistry/stampede/XR_sfr_1e-2/heattime_0201.npz')
hJ100 = np.load(os.getenv('HOME')+'/sim/chemistry/stampede/XR_sfr_1e-1/heattime_0200.npz')
```

```python
fig = plt.figure(2, figsize=(16,12))
ax0 = fig.add_subplot(221)
ax1 = fig.add_subplot(222)
ax2 = fig.add_subplot(223)
ax3 = fig.add_subplot(224)

ax0.hexbin(V['n'][::s],V['tchem'][::s]/yr, gridsize=gs, bins='log', xscale='log', yscale='log', mincnt=1, cmap=cool)
ax1.hexbin(J0['n'][::s],J0['tchem'][::s]/yr, gridsize=gs, bins='log', xscale='log', yscale='log', mincnt=1, cmap=cool)
ax2.hexbin(J10['n'][::s],J10['tchem'][::s]/yr, gridsize=gs, bins='log', xscale='log', yscale='log', mincnt=1, cmap=cool)
ax3.hexbin(J100['n'][::s],J100['tchem'][::s]/yr, gridsize=gs, bins='log', xscale='log', yscale='log', mincnt=1, cmap=cool)


ax0.hexbin(hV['n'],hV['tchem']/yr, gridsize=80, bins='log', xscale='log', yscale='log', mincnt=1, cmap=warm)
ax1.hexbin(hJ0['n'][::s],hJ0['tchem'][::s]/yr, gridsize=gs, bins='log', xscale='log', yscale='log', mincnt=1, cmap=warm)
ax2.hexbin(hJ10['n'][::s],hJ10['tchem'][::s]/yr, gridsize=gs, bins='log', xscale='log', yscale='log', mincnt=1, cmap=warm)
ax3.hexbin(hJ100['n'][::s],hJ100['tchem'][::s]/yr, gridsize=gs, bins='log', xscale='log', yscale='log', mincnt=1, cmap=warm)

ndens = np.logspace(-3, 12, 100)
rho = 1.6726e-24 * ndens
tff = np.sqrt(3 / (32 * np.pi * 6.67e-8 * rho))

ax0.plot(ndens,tff/yr, 'k', lw=2.5, label='freefall time')
ax1.plot(ndens,tff/yr, 'k', lw=2.5, label='freefall time')
ax2.plot(ndens,tff/yr, 'k', lw=2.5, label='freefall time')
ax3.plot(ndens,tff/yr, 'k', lw=2.5, label='freefall time')

ax0.set_xlim(2e-1,1e12)
ax1.set_xlim(2e-1,1e12)
ax2.set_xlim(2e-1,1e12)
ax3.set_xlim(2e-1,1e12)

ax0.set_ylim(20,1e10)
ax1.set_ylim(20,1e10)
ax2.set_ylim(20,1e10)
ax3.set_ylim(20,1e10)

ax0.xaxis.set_visible(False)
ax1.xaxis.set_visible(False)
ax1.yaxis.set_visible(False)
ax3.yaxis.set_visible(False)
ax2.set_xticks((1e0,1e2,1e4,1e6,1e8,1e10))
ax3.set_xticks((1e0,1e2,1e4,1e6,1e8,1e10,1e12))
ax0.set_yticks((1e2,1e4,1e6,1e8,1e10))
ax2.set_yticks((1e2,1e4,1e6,1e8))

ax0.set_ylabel('Heating Time [yr]')
ax2.set_ylabel('Heating Time [yr]')
ax2.set_xlabel('Number Density [cm$^{-3}$]')
ax3.set_xlabel('Number Density [cm$^{-3}$]')

ax0.text(.97,.97, r'J = 0', ha='right', va='top', size=25, transform=ax0.transAxes)
ax1.text(.97,.97, r'J = J$_{0}$', ha='right', va='top', size=25, transform=ax1.transAxes)
ax2.text(.97,.97, r'J = 10 J$_{0}$', ha='right', va='top', size=25, transform=ax2.transAxes)
ax3.text(.97,.97, r'J = 100 J$_{0}$', ha='right', va='top', size=25, transform=ax3.transAxes)

plt.tight_layout(w_pad=0.0,h_pad=0.0)
fig.savefig('figures/dtchem/heattime_t0.png', bbox_inches='tight')
```
