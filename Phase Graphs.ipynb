# The First Stars: Formation under Cosmic Ray Feedback -- Phase Diagrams

This notebook generates all simulation results figures for my current paper.  All simulations run on stampede.tacc.utexas.edu

```python
%matplotlib inline
import os
import numpy as np
import pandas as pd
from astropy.cosmology import WMAP9
import matplotlib as mpl
from matplotlib import pyplot as plt
#mpl.rc('font', size=20.)
#mpl.rc('font', family='serif')
#mpl.rc('text', usetex=True)
import seaborn as sns
import pyGadget
```

```python
sns.set_context('poster')
mpl.rc('text', usetex=True)
```

```python
simV = pyGadget.sim.Simulation('stampede/vanilla',length='pc')
sim0 = pyGadget.sim.Simulation('stampede/xr_tau_J0',length='pc')
sim1 = pyGadget.sim.Simulation('stampede/xr_tau_J1',length='pc')
sim2 = pyGadget.sim.Simulation('stampede/xr_tau_J2',length='pc')
sim3 = pyGadget.sim.Simulation('stampede/xr_tau_J3',length='pc')
sims = [simV, sim0, sim1, sim2, sim3]
```

```python
simname = {'vanilla':r'J = 0',
        'xr_tau_J0': r'J = J$_{0}$', 
        'xr_tau_J1': r'J = 10 J$_{0}$', 
        'xr_tau_J2': r'J = 10$^2$ J$_{0}$',
        'xr_tau_J3': r'J = 10$^3$ J$_{0}$'}
simlist = ['vanilla', 'xr_tau_J0', 'xr_tau_J1', 'xr_tau_J2', 'xr_tau_J3']
nsnap = [355, 306, 327, 271, 235]
```

```python
snaps= {}
for name, n, sim in zip(simlist, nsnap, sims):
    s = sim.load_snapshot(n, 'ndensity', 'temp', 'particleIDs', 'electron_frac')
    s.gas.cleanup('ndensity', 'temp', 'particleIDs', 'electron_frac')
    snaps[name] = s
```

```python
snapdata = {}
for sim in simname.keys():
    snap = snaps[sim]
    snapdata[sim] = pd.DataFrame(np.column_stack((snap.gas.particleIDs, snap.gas.ndensity, snap.gas.temp, snap.gas.electron_frac)),
                               index=snap.gas.particleIDs, columns=['ID', 'ndensity', 'temp', 'efrac'])
    snap.gas.cleanup()
```

```python
average = {}
store = pd.HDFStore(os.getenv("HOME")+'/data/simdata/xrays/binned_averages.hdf5')
bins = np.linspace(-4,12.2,82)
for sim in simname.keys():
    print sim
    df = snapdata[sim]
    df['logn'] = np.log10(df.ndensity)
    df = df.sort('ndensity')
    gb = df.groupby(pd.cut(df.logn, bins))
    average[sim] = gb.mean()
    store[sim] = average[sim]
store.close()
```

```python
current_palette = sns.color_palette()[::-1]
sns.palplot(current_palette)
```

```python
current_palette = sns.color_palette()
f,ax = plt.subplots()
df = average['vanilla']
ax.plot(df.logn, np.log10(df.temp), color='.25', label=simname['vanilla'])
for sim in simlist[1:]:
    df = average[sim]
    ax.plot(df.logn, np.log10(df.temp), label=simname[sim])
ax.legend(loc=0)
ax.set_xlim(-1, 12)
ax.set_ylim(1.75,3.25)
ax.set_xlabel('Log n [cm$^{-3}$]')
ax.set_ylabel('Log Temperature [K]')
f.savefig('figures/phase_diagrams/stacy_temp.png', bbox_inches='tight')
```

```python
current_palette = sns.color_palette()[::-1]
f,ax = plt.subplots()
df = average['vanilla']
ax.plot(df.logn, np.log10(df.efrac), color='.25', label=simname['vanilla'])
for sim in simlist[1:]:
    df = average[sim]
    ax.plot(df.logn, np.log10(df.efrac), label=simname[sim])
ax.legend(loc=0)
ax.set_xlim(-1,12)
ax.set_ylim(-10,-2)
ax.set_xlabel('Log n [cm$^{-3}$]')
ax.set_ylabel('Log $f_{e^-}$')
```

```python
with sns.color_palette(n_colors=4):
    f,(ax0,ax1) = plt.subplots(2, sharex=True, figsize=(8,12))
    
ax0.plot(average['vanilla'].logn, np.log10(average['vanilla'].logT, color='.25', label=simname['vanilla'])
for sim in simlist[1:]:
    df = average[sim]
    ax0.plot(df.logn, df.logT, label=simname[sim])
ax0.axvline(-.7, color='.5', alpha=.3)
ax0.set_ylim(1.,4.2)
ax0.set_ylabel('Log Temperature [K]')

ax1.plot(average['vanilla'].logn, np.log10(average['vanilla'].loge, color='.25', label=simname['vanilla'])
for sim in simlist[1:]:
    df = average[sim]
    ax1.plot(df.logn, df.loge, label=simname[sim])
ax1.axvline(-.7, color='.5', alpha=.3)
ax1.legend(loc=0)
#ax1.set_ylim(-10,-8)
ax1.set_ylabel("Log free $e^-$ fraction")


ax1.set_xlim(-2.5,8)
ax1.set_xlabel("Log number density [cm$^{-3}$]")
f.subplots_adjust(hspace=0.02)
f.savefig('figures/phase_diagrams/binned.png', bbox_inches='tight')
```
