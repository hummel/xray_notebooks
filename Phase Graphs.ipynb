# The First Stars: Formation under Cosmic Ray Feedback -- Phase Diagrams

This notebook generates all simulation results figures for my current paper.  All simulations run on stampede.tacc.utexas.edu

```python
%matplotlib inline
import os
import numpy as np
import pandas as pd
import matplotlib as mpl
from matplotlib import pyplot as plt
#mpl.rc('font', size=20.)
#mpl.rc('font', family='serif')
#mpl.rc('text', usetex=True)
import seaborn as sns
import pyGadget
```

```python
sns.set_context('poster')
mpl.rc('text', usetex=True)
```

```python
simV = pyGadget.sim.Simulation('stampede/vanilla',length='pc')
sim0 = pyGadget.sim.Simulation('stampede/xr_tau_J0',length='pc')
sim1 = pyGadget.sim.Simulation('stampede/xr_tau_J1',length='pc')
sim2 = pyGadget.sim.Simulation('stampede/xr_tau_J2',length='pc')
sim3 = pyGadget.sim.Simulation('stampede/xr_tau_J3',length='pc')
sims = [simV, sim0, sim1, sim2, sim3]
```

```python
simname = {'vanilla':r'J = 0',
        'xr_tau_J0': r'J = J$_{0}$', 
        'xr_tau_J1': r'J = 10 J$_{0}$', 
        'xr_tau_J2': r'J = 10$^2$ J$_{0}$',
        'xr_tau_J3': r'J = 10$^3$ J$_{0}$'}
simlist = ['vanilla', 'xr_tau_J0', 'xr_tau_J1', 'xr_tau_J2', 'xr_tau_J3']
nsnap = [355, 306, 327, 271, 235]
```

```python
snaps= {}
for name, n, sim in zip(simlist, nsnap, sims):
    s = sim.load_snapshot(n, 'ndensity', 'temp', 'particleIDs', 'electron_frac')
    s.gas.cleanup('ndensity', 'temp', 'particleIDs', 'electron_frac')
    snaps[name] = s
```

```python
snapdata = {}
for sim in simname.keys():
    snap = snaps[sim]
    snapdata[sim] = pd.DataFrame(np.column_stack((snap.gas.particleIDs, snap.gas.ndensity, snap.gas.temp, snap.gas.electron_frac)),
                               index=snap.gas.particleIDs, columns=['ID', 'ndensity', 'temp', 'efrac'])
    snap.gas.cleanup()
```

```python
average = {}
bins = np.linspace(-4,12.2,82)
for sim in simname.keys():
    print sim
    df = snapdata[sim]
    df['logn'] = np.log10(df.ndensity)
    df['logT'] = np.log10(df.temp)
    df['loge'] = np.log10(df.efrac)
    df = df.sort('ndensity')
    gb = df.groupby(pd.cut(df.logn, bins))
    average[sim] = gb.mean()
```

```python
current_palette = sns.color_palette()[::-1]
sns.palplot(current_palette)
```

```python
current_palette = sns.color_palette()
f,ax = plt.subplots()
df = average['vanilla']
ax.plot(df.logn, df.logT, color='.25', label=simname['vanilla'])
for sim in simlist[1:]:
    df = average[sim]
    ax.plot(df.logn, df.logT, label=simname[sim])
ax.legend(loc=0)
ax.set_xlim(-1, 12)
ax.set_ylim(1.75,3.25)
ax.set_xlabel('Log n [cm$^{-3}$]')
ax.set_ylabel('Log Temperature [K]')
f.savefig('figures/phase_diagrams/stacy_temp.png', bbox_inches='tight')
```

```python
current_palette = sns.color_palette()[::-1]
f,ax = plt.subplots()
df = average['vanilla']
ax.plot(df.logn, df.loge, color='.25', label=simname['vanilla'])
for sim in simlist[1:]:
    df = average[sim]
    ax.plot(df.logn, df.loge, label=simname[sim])
ax.legend(loc=0)
ax.set_xlim(-1,12)
ax.set_ylim(-10,-2)
ax.set_xlabel('Log n [cm$^{-3}$]')
ax.set_ylabel('Log $f_{e^-}$')
```

```python
f,(ax0,ax1) = plt.subplots(2, sharex=True, figsize=(8,12))
ax0.plot(average['vanilla'].logn, average['vanilla'].logT, color='.25', label=simname['vanilla'])
for sim in simlist[1:]:
    df = average[sim]
    ax0.plot(df.logn, df.logT, label=simname[sim])
ax0.legend(loc=0)
ax0.set_ylabel('Log Temperature [K]')

ax1.plot(average['vanilla'].logn, average['vanilla'].loge, color='.25', label=simname['vanilla'])
for sim in simlist[1:]:
    df = average[sim]
    ax1.plot(df.logn, df.loge, label=simname[sim])
#ax1.legend(loc=0)
#ax1.set_xlim(8,12)
#ax1.set_ylim(-10,-8)
ax1.set_ylabel("Log free $e^-$ fraction")
ax1.set_xlabel("Log number density [cm$^{-3}$]")
f.subplots_adjust(hspace=0.02)
f.savefig('figures/phase_diagrams/binned.png', bbox_inches='tight')
```

# Ionization vs Recombination:

```python
bkg = {'xr_tau_J0': 1e0, 'xr_tau_J1': 1e1, 'xr_tau_J2': 1e2, 'xr_tau_J3': 1e3}
```

```python
def find_closest(A, target):
    #A must be sorted
    idx = A.searchsorted(target)
    idx = np.clip(idx, 1, len(A)-1)
    left = A[idx-1]
    right = A[idx]
    idx -= target - left < right - target
    return idx
```

```python
jxr = pd.read_csv('jxr.dat', delim_whitespace=True, names=['z', 'J0'])
jxr.sort('z', inplace=True)
```

```python
kh_rates = pd.read_csv('kh_rates.dat', delim_whitespace=True,
                       names=['logn', 'kH', 'kHe', 'kHeII', 'hH', 'hHe', 'hHeII'])
kH = kh_rates[['logn', 'kH']]
for sim in simlist[1:]:
    snap = snaps[sim]
    z, J0 = jxr.iloc[find_closest(jxr.z.values, snap.header.Redshift)]
    print sim, snap.header.Redshift, J0
    kH[sim] = kH.kH*J0*bkg[sim]
```

```python
for sim in simlist:
    df = average[sim]
    df['n'] = bins[1:]
    df['k_rec'] = df.ndensity**2 * df.efrac**2 * 2.6e-13 * (df.temp/1e4)**-.75
```

```python
for sim in simlist:
    snap = snaps[sim]
    z, J0 = jxr.iloc[find_closest(jxr.z.values, snap.header.Redshift)]
    print sim, snap.sim.name, snap.header.Redshift, z, J0
```

```python
with sns.color_palette(n_colors=4):
    f,ax = plt.subplots()
df = average['vanilla']
ax.semilogy(df.logn, df.k_rec, color='.25', label=simname['vanilla'])
for sim in simlist[1:]:
    df = average[sim]
    ax.plot(df.logn, df.k_rec, label=simname[sim])
for sim in simlist[1:]:
    ax.plot(kH.logn, kH[sim], ls='--')

ax.legend(loc=0)
ax.set_xlim(-4, 6)
ax.set_ylim(1e-30, 1e-14)
ax.set_xlabel('Log n [cm$^{-3}$]')
ax.set_ylabel('Ionisation / Recombination Rate [$s^{-1}$]')
#f.savefig('figures/phase_diagrams/stacy_temp.png', bbox_inches='tight')
```
