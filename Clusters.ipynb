```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib as mpl
from matplotlib import pyplot as plt
mpl.rc('figure', figsize=(12,9))
mpl.rc('font', size=20.)

mpl.rc('font', family='serif')
mpl.rc('text', usetex=True)
mpl.rc('axes', color_cycle=['k','b','g','r','yellow'])

import pyGadget
```

```python
from scipy.cluster.vq import kmeans,vq
```

```python
sim0 = pyGadget.sim.Simulation('stampede/vanilla')
sim1 = pyGadget.sim.Simulation('stampede/XR_sfr_1e-3')
sim2 = pyGadget.sim.Simulation('stampede/XR_sfr_1e-2')
sim3 = pyGadget.sim.Simulation('stampede/XR_sfr_1e-1')
```

```python
snap = sim2.load_snapshot(1701, 'ndensity', 'coordinates', 'velocities')
```

```python
xyz = snap.gas.get_coords('AU',centering='avg')
uvw = snap.gas.get_velocities()
dens = snap.gas.get_number_density()
temp = snap.gas.get_temperature()
```

```python
xyz, uvw, dens, temp = pyGadget.analyze.data_slice(snap.gas.ndensity > 1e8, snap.gas.coordinates, snap.gas.velocities,
                                                   snap.gas.ndensity, snap.gas.temp)
```

```python
centroids,_ = kmeans(xyz, 4)
idx,_ = vq(xyz, centroids)
```

```python
plt.plot(-xyz[idx==0,0],-xyz[idx==0,1],'ok',
         -xyz[idx==1,0],-xyz[idx==1,1],'ob',
         -xyz[idx==2,0],-xyz[idx==2,1],'og',
         -xyz[idx==3,0],-xyz[idx==3,1],'or')
plt.xlabel('AU')
plt.ylabel('AU')
plt.savefig('figures/clusters/clustering.png')
```

```python
fig = plt.figure(2, figsize=(16,12))
ax0 = fig.add_subplot(221)
ax1 = fig.add_subplot(222)
ax2 = fig.add_subplot(223)
ax3 = fig.add_subplot(224)

color = ['k', 'b', 'g', 'r']
for i,ax in enumerate([ax0, ax1, ax2, ax3]):
    ax.scatter(dens[idx==i], temp[idx==i], c=color[i])
    ax.set_xscale('log')
    ax.set_yscale('log')
    ax.set_xlim(1e8, 1e12)
    ax.set_ylim(10, 1e4)
    ax.axhline(2.725 * (snap.header.Redshift + 1),
               linestyle='--', color='k')
    ax.set_xlabel('n [cm$^{-3}$]')
    ax.set_ylabel('Temperature [K]')

ax0.xaxis.set_visible(False)
ax1.xaxis.set_visible(False)
ax1.yaxis.set_visible(False)
ax3.yaxis.set_visible(False)
ax2.set_xticks((1e8,1e9,1e10,1e11))
ax3.set_xticks((1e9,1e10,1e11,1e12))
ax0.set_yticks((1e2,1e3,1e4))
ax2.set_yticks((10,1e2,1e3))

#ax0.text(.97,.97, r'J = 0', ha='right', va='top', size=25, transform=ax0.transAxes)
#ax1.text(.97,.97, r'J = J$_{0}$', ha='right', va='top', size=25, transform=ax1.transAxes)
#ax2.text(.97,.97, r'J = 10 J$_{0}$', ha='right', va='top', size=25, transform=ax2.transAxes)
#ax3.text(.97,.97, r'J = 100 J$_{0}$', ha='right', va='top', size=25, transform=ax3.transAxes)

plt.tight_layout(w_pad=0.0,h_pad=0.0)
fig.savefig('figures/clusters/temp.png', bbox_inches='tight')
```
