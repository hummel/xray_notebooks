```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib as mpl
from matplotlib import pyplot as plt
mpl.rc('font', size=20.)
mpl.rc('font', family='serif')
mpl.rc('text', usetex=True)
import pyGadget
```

```python
sim = pyGadget.sim.Simulation('stampede/vanilla',length='AU')
```

```python
snap = sim.load_snapshot(355)
```

```python
n = snap.gas.get_number_density()
T = snap.gas.get_temperature()
snap.gas.calculate_optical_depth(1.14093e-23)
opd = snap.gas.tau
n,T,opd = pyGadget.analyze.data_slice(n >= 1e4, n,T,opd)
```

```python
fig = plt.figure(1, figsize=(16,12))
ax = fig.add_subplot(111)
ax.set_xscale('log')
ax.set_yscale('log')
im = ax.scatter(n,T, c=np.log10(opd), cmap=plt.cm.seismic, linewidths=0.)
im.set_clim(-1,1)
cb = plt.colorbar(im)
ax.set_xlim(1e4,1.5e12)

ax.set_xlabel('n [cm$^{-3}$]')
ax.set_ylabel('Temperature [K]')
cb.set_label('Log optical depth')

ax.text(.05,.95, r'1 keV', ha='left', va='top', size=25, transform=ax.transAxes)
plt.savefig('figures/opd-1keV.png', bbox_inches='tight')
```
