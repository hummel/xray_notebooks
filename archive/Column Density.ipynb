```python
%matplotlib inline
import numpy as np
import pandas as pd
from scipy import optimize
from scipy.interpolate import griddata
import matplotlib as mpl
from matplotlib import pyplot as plt
#import seaborn as sns
mpl.rcParams['figure.figsize'] = (16,11)
mpl.rcParams['font.size'] =20
mpl.rc('font', family='serif')
mpl.rc('text', usetex=True)
import pyGadget
```

```python
sim = pyGadget.sim.Simulation('stampede/vanilla',length='AU')
```

```python
snap = sim.load_snapshot(355)
snap.gas.load_coords()
snap.gas.orient_box(centering='max')
```

```python
r2d = np.linalg.norm(snap.gas.coordinates[:,:2], axis=1)

rad1 = 50
rad2 = 5000
cube = 2000
pos1, n1 = pyGadget.analyze.data_slice(r2d < rad1, snap.gas.coordinates, snap.gas.ndensity)
#pos2, n2 = pyGadget.analyze.data_slice(r2d < rad2, snap.gas.coordinates, snap.gas.ndensity)
pos2, n2 = pyGadget.analyze.data_slice((np.abs(snap.gas.coordinates[:,0]) <= cube) 
                                       & (np.abs(snap.gas.coordinates[:,1]) <= cube),
                                       snap.gas.coordinates, snap.gas.ndensity)
plt.loglog(np.abs(pos2[:,2]), n2, 'bo')
plt.loglog(np.abs(pos1[:,2]), n1, 'ro')
plt.axhline(1e4)
```

```python
r1 = np.linalg.norm(pos1, axis=1)
print r1.size, r1.min(), r1.max()
r2 = np.linalg.norm(pos2, axis=1)
print r2.size, r2.min(), r2.max()
```

```python
bins = np.logspace(-2,np.log10(2e5), 100)
r = r2
n = n2
```

```python
r.min()
```

```python
r[r < 1e-2] = 1e-2
binned_n = []
midpoints = []
for i in range(bins.size-1):
    mid = (bins[i] + bins[i+1]) / 2
    in_bin = np.where((r>=bins[i]) & (r<bins[i+1]))[0]
    if in_bin.size > 1:
        avg = n[in_bin].mean()
    elif in_bin.size == 1:
        avg = n[in_bin][0]
    else:
        if mid > 1e2:
            avg = np.nan
    #print mid, avg
    midpoints.append(mid)
    binned_n.append(avg)
midpoints = np.asarray(midpoints)
binned_n = np.asarray(binned_n)
```

```python
plt.loglog(r, n, 'ko', ms = 2)
plt.plot(midpoints,binned_n, 'ro', ms=5)
plt.plot(bins,1e15 * bins**(-2), 'b--', lw=3)
plt.axhline(1e4)

plt.xlim(.009, 1e6)
#plt.ylim(1e7, 1e13)
```

```python
bins.size, binned_n.size
```

```python
bins[0] = 0.0
N = 0.0
NList = []
for i in range(bins.size-1, 0, -1):
    #print i, bins[i], bins[i-1], binned_n[i-1]
    N += binned_n[i-1] * (bins[i] - bins[i-1]) * 1.49597871e13
    NList.append(N)
N = np.asarray(NList)
```

```python
sigma = 1.14e-23
optical_depth = N*sigma
optical_depth[0], optical_depth[-1]
```

```python
plt.loglog(midpoints, N[::-1]*sigma)
plt.xlabel('Radius [AU]')
plt.ylabel('Optical Depth')
```

```python
plt.loglog(binned_n[::-1], N*sigma)
plt.xlim(1e12, 5e3)
plt.xlabel('Number Density [cm$^{-3}$]')
plt.ylabel('Optical Depth')
```
