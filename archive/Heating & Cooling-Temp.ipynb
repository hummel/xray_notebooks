```python
%matplotlib inline
import os
import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt
mpl.rc('figure', figsize=(16,12))
mpl.rc('font', size=20.)
mpl.rc('font', family='serif')
mpl.rc('text', usetex=True)
```

# Vanilla

### At Sink Formation

```python
cool = np.load(os.getenv('HOME')+'/data/chemistry/stampede/vanilla/cooltime_0355.npz')
heat = np.load(os.getenv('HOME')+'/data/chemistry/stampede/vanilla/heattime_0355.npz')
plt.hexbin(cool['n'],cool['temp'], gridsize=500, bins='log', xscale='log', yscale='log', mincnt=1, cmap=plt.cm.Blues_r)
plt.hexbin(heat['n'],heat['temp'], gridsize=80, bins='log', xscale='log', yscale='log', mincnt=1, cmap=plt.cm.autumn)
plt.xlim(2e-3,1e12)
plt.ylim(10, 2e4)
plt.xlabel('n [cm$^{-3}$]')
plt.ylabel('Temperature [K]')
plt.savefig('figures/temp/vanilla_t0.png', bbox_inches='tight')
```
