```python
%matplotlib inline
import os
import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt
mpl.rc('figure', figsize=(16,12))
mpl.rc('font', size=20.)
mpl.rc('font', family='serif')
mpl.rc('text', usetex=True)
yr = 3600 * 24 * 365.25
```

# Vanilla

### At Sink Formation

```python
cool = np.load(os.getenv('HOME')+'/data/chemistry/stampede/vanilla/cooltime_0355.npz')
heat = np.load(os.getenv('HOME')+'/data/chemistry/stampede/vanilla/heattime_0355.npz')
plt.hexbin(cool['n'],cool['tchem']/yr, gridsize=500, bins='log', xscale='log', yscale='log', mincnt=1, cmap=plt.cm.Blues_r, label="Cooling")
plt.hexbin(heat['n'],heat['tchem']/yr, gridsize=80, bins='log', xscale='log', yscale='log', mincnt=1, cmap=plt.cm.autumn, label="Heating")
ndens = np.logspace(-3, 12, 100)
rho = 1.6726e-24 * ndens
tff = np.sqrt(3 / (32 * np.pi * 6.67e-8 * rho))
plt.plot(ndens,tff/yr, 'k', lw=2.5, label='freefall time')
plt.xlim(2e-3,1e12)
plt.ylim(50,5e13)
plt.xlabel('n [cm$^{-3}$]')
plt.ylabel('Heating / Cooling Time [yr]')
plt.legend()
plt.savefig('figures/dtchem/vanilla_t0.png', bbox_inches='tight')
```
