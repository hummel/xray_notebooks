```python
%matplotlib inline
import numpy as np
from scipy.interpolate import griddata
import seaborn as sns
from matplotlib import pyplot as plt
import pyGadget
```

```python
sns.set_context('poster')
```

```python
sim = pyGadget.sim.Simulation('stampede/vanilla',length='AU')
```

```python
s1 = sim.load_snapshot(355)
s1.gas.load_coords()
s1.gas.orient_box(centering='max', view='face')
```

```python
s2 = sim.load_snapshot(615)
s2.gas.load_coords()
s2.gas.orient_box(centering='max', view='face')
```

```python
r2d1 = np.linalg.norm(s1.gas.coordinates[:,:2], axis=1)
r2d2 = np.linalg.norm(s2.gas.coordinates[:,:2], axis=1)

rad0 = 50
rad = 1e5
pos1_0, n1_0 = pyGadget.analyze.data_slice(r2d1 < rad0, s1.gas.coordinates, s1.gas.ndensity)
pos1, n1 = pyGadget.analyze.data_slice(r2d1 < rad, s1.gas.coordinates, s1.gas.ndensity)
pos2_0, n2_0 = pyGadget.analyze.data_slice(r2d2 < rad0, s2.gas.coordinates, s2.gas.ndensity)
pos2, n2 = pyGadget.analyze.data_slice(r2d2 < rad, s2.gas.coordinates, s2.gas.ndensity)
```

```python
plt.loglog(np.abs(pos1[:,2]), n1, 'o', ms=2)
plt.loglog(np.abs(pos1_0[:,2]), n1_0, 'o', ms=5)
plt.loglog(np.abs(pos2[:,2]), n2, 'o', ms=2)
plt.loglog(np.abs(pos2_0[:,2]), n2_0, 'o', ms=1)

plt.xlim(1e-2,1e8)
plt.ylim(1e-3,1.5e12)
plt.xlabel('z-Height [AU]')
plt.ylabel('Number density [cm$^{-3}$]')
```

```python
s1.gas.cleanup()
s1.close()
s2.gas.cleanup()
s2.close()
```

# Interpolating Column with griddata

```python
xmax = ymax = 1
xmin = ymin = -1
zmax = 1e6
zmin = 0
(zmax-zmin)
```

```python
grid_x, grid_y, grid_z = np.mgrid[xmin:xmax:3j, ymin:ymax:3j, zmin:zmax:1000001j]
```

```python
grid_z.shape, grid_z.size
```

```python
grid_z[0,0,1] - grid_z[0,0,0] #Double-check grid spacing!
```

```python
grid_n1 = griddata(pos1, n1, (grid_x, grid_y, grid_z), method='linear')
grid_n2 = griddata(pos2, n2, (grid_x, grid_y, grid_z), method='linear')
```

```python
plt.loglog(np.abs(pos1[:,2]), n1, 'o', ms=3)
plt.loglog(np.abs(pos2_0[:,2]), n2_0, 'o', ms=3)
plt.plot(grid_z[1,1], grid_n1[1,1])
plt.plot(grid_z[1,1], grid_n2[1,1])
```

```python
z = grid_z[1,1]
nz1 = grid_n1[1,1]
nz2 = grid_n2[1,1]
sigma = 1.14e-23
dz = z[1] - z[0]
dz
```

```python
N1 = nz1[::-1].cumsum()
N1 = N1[::-1] * 1.49597871e13
tau1 = N1 * sigma

N2 = nz2[::-1].cumsum()
N2 = N2[::-1] * 1.49597871e13
tau2 = N2 * sigma

tau1[0], tau2[0]
```

```python
plt.loglog(z, tau1)
plt.loglog(z, tau2)
plt.xlim(1, 1e7)
plt.ylim(1e-5, 1e5)
plt.xlabel('Radius [AU]')
plt.ylabel('Optical Depth')
```

```python
plt.loglog(nz1, tau1)
plt.loglog(nz2, tau2)
plt.xlim(1e12, .1)
plt.ylim(1e-5, 1e4)
plt.xlabel('Number Density [cm$^{-3}$]')
plt.ylabel('Optical Depth')
```
