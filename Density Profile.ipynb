```python
%matplotlib inline
import numpy as np
import pandas as pd
from scipy import optimize
import matplotlib as mpl
from matplotlib import pyplot as plt
mpl.rc('figure', figsize=(12,9))
mpl.rc('font', size=20.)
mpl.rc('font', family='serif')
mpl.rc('text', usetex=True)
import pyGadget
```

```python
sim = pyGadget.sim.Simulation('stampede/vanilla',length='AU')
```

```python
snap = sim.load_snapshot(355)
snap.gas.load_coords()
snap.gas.orient_box(centering='max')
```

```python
cylinder = np.linalg.norm(snap.gas.coordinates[:,:2], axis=1)
#sphere = np.linalg.norm(snap.gas.coordinates, axis=1)

size = 5000
pos, n = pyGadget.analyze.data_slice(cylinder < size, 
                                     snap.gas.coordinates, snap.gas.ndensity)
plt.loglog(np.abs(pos[:,2]), n, 'bo')
```

```python
r = np.linalg.norm(pos, axis=1)
print r.size, r.min(), r.max()
```

```python
plt.loglog(np.abs(pos[:,2]), n, 'bo')
plt.loglog(np.abs(pos[close,2]), n[close], 'ro')
plt.loglog(r, n, 'go')
plt.loglog(r[close], n[close], 'co')
plt.xlim(.1, 1e4)
plt.ylim(1e7, 1e13)
```

```python
bins = np.logspace(-2,7, 100)
r[n.argmax()] = 1e-2
```

```python
#plt.loglog(np.abs(pos[:,2]), n, 'bo')
#plt.loglog(np.abs(pos[close,2]), n[close], 'ro')
plt.loglog(r, n, 'go')
plt.loglog(r[close], n[close], 'co')

plt.plot(bins,1e15 * bins**(-2), 'r', lw=3)
for x in bins:
    plt.axvline(x)

plt.xlim(.1, 1e4)
plt.ylim(1e7, 1e13)
```

```python
binned_n = []
midpoints = []
for i in range(1,bins.size):
    mid = (bins[i-1] + bins[i]) / 2
    in_bin = np.where((r>=bins[i-1]) & (r<bins[i]))[0]
    if in_bin.size > 1:
        avg = n[in_bin].mean()
    elif in_bin.size == 1:
        avg = n[in_bin][0]
    else:
        if mid > 1e2:
            avg = np.nan
    #print mid, avg
    midpoints.append(mid)
    binned_n.append(avg)
midpoints = np.asarray(midpoints)
binned_n = np.asarray(binned_n)
```

```python
plt.loglog(r, n, 'ko', ms = 2)
plt.plot(midpoints,binned_n, 'ro', ms=5)
plt.plot(bins,1e15 * bins**(-2), 'b--', lw=3)

#plt.plot(10**x, 10**y, 'r:', lw=3)

#plt.xlim(.009, 1e4)
#plt.ylim(1e7, 1e13)
```

```python
tail = np.where((n < 5e11))[0]
```

```python
log_r = np.log10(r[tail])
log_n = np.log10(n[tail])
log_r, log_n
```

```python
def f(x, A, B):
    return A*x + B
```

```python
(A,B),err = optimize.curve_fit(f, log_r, log_n)
x = np.linspace(1,5, 20)
A,B,err
```

```python
y = f(x, A,B)
x, y
```
